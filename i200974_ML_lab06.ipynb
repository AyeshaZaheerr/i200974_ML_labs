```python
def naivebayes(f1, f2, f3, f4):
    no_of_outlook = 0 #count of no's in the first column 
    no_of_temp = 0 #count of no's in the 2nd column 
    no_of_humidity = 0 #count of no's in the third column 
    no_of_wind = 0 #count of no's in the 4th column 
    no_of_observations=len(df['outlook'])
    
    yes_of_outlook = 0 #count of yes's in the first column 
    yes_of_temp = 0 #count of yes's in the 2nd column 
    yes_of_humidity = 0 #count of yes's in the third column 
    yes_of_wind = 0 #count of yes in the 4th column 
    
  
    for i in range (0,no_of_observations):
        if ((df['outlook'][i] == f1) and (df['play'][i] == 'No')):
            no_of_outlook=no_of_outlook+1
            
    for i in range (0, len(df['outlook'])):
        if ((df['outlook'][i] == f1) and (df['play'][i] == 'Yes')):
            yes_of_outlook=yes_of_outlook+1
   
    for i in range (0,no_of_observations):
        if ((df['temp'][i] == f2)   and (df['play'][i] == 'No')):
            no_of_temp=no_of_temp+1
            
    for i in range (0, len(df['temp'])):
        if ((df['temp'][i] == f2)   and (df['play'][i] == 'Yes')):
            yes_of_temp=yes_of_temp+1
    
    for i in range (0,no_of_observations):
        if ((df['humidity'][i] == f3)   and (df['play'][i] == 'No')):
            no_of_humidity=no_of_humidity+1
            
    for i in range (0, len(df['humidity'])):
        if ((df['humidity'][i] == f3)   and (df['play'][i] == 'Yes')):
            yes_of_humidity=yes_of_humidity+1
   
    for i in range (0,no_of_observations):
        if ((df['wind'][i] == f4)   and (df['play'][i] == 'No')):
            no_of_wind=no_of_wind+1
    
    for i in range (0, len(df['wind'])):  
        if ((df['wind'][i] == f4)   and (df['play'][i] == 'Yes')):
            yes_of_wind=yes_of_wind+1
            
    #counting the total number of yes and nos respectively 
    yes=0
    no=0
    play_col=df['play']
    for i in play_col:
        if(i=="Yes"):
            yes=yes+1
        else:
            no=no+1

    proby = ((yes_of_outlook/yes) * (yes_of_temp/yes) * (yes_of_humidity/yes) * (yes_of_wind/yes) * (yes/no_of_observations))
    print("prob of play=yes ",proby)

    
    probno = ((no_of_outlook/no) * (no_of_temp/no) * (no_of_humidity/no) * (no_of_wind/no) * (no/no_of_observations))
    print("prob of play=no",probno)
    
    if proby > probno:
        print("play=yes")
    else:
        print("play=no")

print("Enter outlook, temp,humidity and wind")
input_outlook=input("outlook :")
input_temp=input("temp:")
input_humidity=input("humidity:")
input_wind=input("wind:")
naivebayes(input_outlook,input_temp,input_humidity,input_wind)
```
