```python
def Entropy(x):
    var=0 #var for sum 
    u,c = np.unique(x , return_counts=True) #unique counts   
    for i in c:
        var1=math.log2(i/(len(x)))
        var2=i/(len(x))
        var =var+ var2 * var1 
    var = var*(-1)
    return var

def entropy(x,y):
    array_yes = []
    array_no  = []
    
    u_y , c_y = np.unique(y , return_counts=True)
    u_x , c_x = np.unique(x , return_counts=True)
    label = Entropy(y)
    
    for j in u_x: 
        yes = 0
        no = 0
        length=len(x)
        for i in range(0,length):
            if j == x[i]:
                if  y[i] == 'Yes':
                    yes=yes +1
                else:
                    no=no+1
        array_no.append(no)
        array_yes.append(yes)   
    sum=0  
    arr=[]
    length=len(u_x)
    for i in range(0,length):
        summ=0
        
        if (array_yes[i]==0):
            var1= math.log2(array_no[i]/(c_x[i]))
            var2= array_no[i]/(c_x[i])
            summ =summ+ -(var2 * var1)
        elif (array_no[i]==0):
            var1=math.log2(array_yes[i]/(c_x[i]))
            var2=array_yes[i]/(c_x[i])
            summ =sum+  -(var2 *  var1)
            
        else:
            var1= math.log2(array_yes[i]/(c_x[i]))
            var2=array_yes[i]/(c_x[i])
            summ =summ+ -(var2 * var1)
            
            var1=math.log2(array_no[i]/(c_x[i]))
            var2=array_no[i]/(c_x[i])
            summ = summ+ -(var2 * var1)
        
        arr.append(summ)
        
    tot=0
    length=len(c_x)
    len1=len(y)
    for i in range(0,length):
        var1=c_x[i]/len1
        var2=arr[i]*var1
        tot =tot+ var2
    
    info_gain = label - tot
   
    print("Info Gain:")
    print(info_gain)
    return info_gain          
```
